```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('train.csv')
df.tail(50)
```

# 1_Box Plot

```python
sns.boxplot(x=df["Age"],color='plum',width=0.2)
plt.grid()
plt.show()
```

### The boxplot interpret that it is positively skewed and here found some outliers.The minimum value start from zero and the maximum value approoximately 64.

# 2_Bar Graph

```python
bar_df=df.groupby('Sex').mean().reset_index()
bar_df
```

```python
plt.bar(bar_df['Sex'],bar_df['Age'])
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Bar chart of Sex vs Age')
plt.show();
```

### The bar chart express difference between female and male.Here found that male is greater than female.

# 3_Pie Chart

```python
df.Sex.value_counts().plot(kind='pie',autopct='%.1f%%')
plt.title('Pie chart of Percentage of Gender');
```

### The pie chart express that the percentage between male and female and notice that the percentage of male is greater than female.

# 4_histogram

```python
sns.histplot(df['Fare'],bins=10,kde=True,color='blue');
plt.title('Histogram of Fare')
```

```python
The histogram express that this distribution is positively skewed.
```

# 5_Scatter plot

```python
plt.scatter(x=df['PassengerId'],y=df['Age'])
plt.title('Relation between Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show();
```

```python
sns.jointplot(x=df["PassengerId"], y=df["Age"], data=df, kind="reg")
plt.show()
```

### This scatter plot express it is positive correlation
