# One Hot Encoding

#### OneHotEncoding by using 

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df={
    'Id':[1,2,3,4,5],
    'Colour':['Red','Blue','Green','Blue','Blue'],
    'Country':['USA','UK','CANADA','USA','USA']
}
df=pd.DataFrame(df)
df
```

```python
df.dtypes
```

```python
df['Colour'].unique()
```

```python
df['Country'].unique()
```

```python
ohe= OneHotEncoder()
```

```python
print(ohe)
```

```python
ohe.fit_transform(df[['Colour','Country']]).toarray()
```

```python
feature_arry=ohe.fit_transform(df[['Colour','Country']]).toarray()
```

```python
ohe.categories_
```

```python
feature_labels= ohe.categories_
```

```python
feature_labels=np.array(feature_labels).ravel()
```

```python
print(feature_labels)
```

```python
pd.DataFrame(feature_arry,columns=feature_labels)
```

```python
features=pd.DataFrame(feature_arry,columns=feature_labels)
```

```python
print(features)
```

```python
new_df=pd.concat([df,features],axis=1) ##Merge to data frame
```

```python
new_df
```

## Dummy Encoding

```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'Element':['Fire','Water','Earth','Air','Fire','Water','Earth','Air','Air','Air']}
            )
print(data,'\n',)
data_encoded=pd.get_dummies(data,drop_first=True)
data_encoded
```

### Effect Encoding

```python
import pandas as pd
import category_encoders as ce
data=pd.DataFrame({'City':['Delhi','Mumbai','Hydrabad','Chennai','Bangalore','Delhi','Hydrabad','Bangalore','Delhi']})
encoder=ce.sum_coding.SumEncoder(cols='City',verbose=False)
print(data,'\n')
encoder.fit_transform(data)
```

### Hash Encoding

```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'Month':['Jan','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec']})
print(data,'\n')
encoder=ce.HashingEncoder(cols='Month',n_components=3)
encoder.fit_transform(data)
```

### Binary Enoding

```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'Month':['Delhi','Mumbai','Hydrabad','Chennai','Bangalore','Delhi','Hydrabad','Bangalore','Delhi']})
print(data,'\n')
data_encoded=ce.BinaryEncoder(cols=['Month'],return_df=True)
data_encoded.fit_transform(data)
data_encoded
```

### BaseNEncoding

```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'City':['Delhi','Mumbai','Hydrabad','Chennai','Bangalore','Delhi','Hydrabad','Bangalore','Delhi']})
print(data,'\n')
data_encoded=ce.BaseNEncoder(cols=['City'],return_df=True,base=5)
data_encoded.fit_transform(data)
data_encoded
```

```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'class':['A','B','C','B','C','A','A','A'],'Marks':[50,30,70,80,45,97,80,68]})
print(data,'\n')
data_encoded=ce.TargetEncoder(cols='class')
data_encoded.fit_transform(data['class'],data['Marks'])
```
